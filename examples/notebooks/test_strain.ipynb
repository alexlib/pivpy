```python
def test_strain():
    """tests shear estimate"""
    data = io.create_sample_field(rows=2, cols=2, noise_sigma=0.0)
    data = data.piv.strain()
    print(data["u"])
    print(data["v"])
    print(data["w"])
    assert np.allclose(data["w"].values, 0.11328125, 1e-6)
```
