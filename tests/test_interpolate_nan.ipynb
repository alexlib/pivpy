```python
import xarray as xr
from pivpy import io, pivpy, graphics
import numpy as np
```

```python
ds = io.create_sample_Dataset(n_frames=1,rows=5,cols=5,noise_sigma=0.5)
```

```python
ds.piv.mean_flow.piv.showf(colorbar=True)
```

```python
ds["u"][1:4,1:4] = np.nan
ds.sel(t=0)["u"].plot()
```

```python
# see https://docs.xarray.dev/en/stable/user-guide/interpolation.html#interpolating-arrays-with-nan
filled_ds = ds.copy()
filled["u"] = ds["u"].interpolate_na(dim=("x"),method='linear')
filled["u"] = filled.interpolate_na(dim=("y"),method='linear')
filled.sel(t=0)["u"].plot()
```

```python
filled_ds = ds.copy()
filled["u"] = ds["u"].interpolate_na(dim=("x"),method='nearest')
filled["u"] = filled.interpolate_na(dim=("y"),method='nearest')
filled.sel(t=0)["u"].plot()
```
